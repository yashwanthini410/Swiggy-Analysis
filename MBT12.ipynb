###Major Data Types



QUALITATIVE DATA
1. **Ordinal Data**:
    - **Definition**: Ordinal data is a type of categorical data with a natural order or ranking between the categories. The order is meaningful, but the differences between the categories may not be consistent or quantifiable.
    - **Example**: Grades (A, B, C, D, F) in a class. While there is a clear order (A is better than B, and so on), the difference between each grade may not be the same.
2. **Nominal Data**:
    - **Definition**: Nominal data is also a type of categorical data, but there is no natural order or ranking between the categories. Each category is distinct and unrelated to the others.
    - **Example**: Colors of cars (red, blue, green). There is no inherent order or ranking among colors; they are simply different categories.

QUANTITATIVE DATA
1. **Discrete Data**:
    - **Definition**: Discrete data consists of separate, distinct values that cannot be broken down into smaller units. These values are often countable and finite.
    - **Example**: Number of students in a class (1, 2, 3, ...). You can't have a fraction or a non-integer number of students; it must be a whole number.
2. **Continuous Data**:
    - **Definition**: Continuous data represents measurements that can take on any value within a range. These values are not countable and can be infinitely divided into smaller units.
    - **Example**: Height of students. Heights can take on any value within a range (e.g., 150 cm, 150.5 cm, 151 cm), and there is no limit to how precise the measurement can be.

## Searching and filtering the data

### - **VLOOKUP**
    - Used to look up values in vertical order
    - It looks up for the first value of selected table and gives the value of selected column index
    
    formula : =VLOOKUP(`INPUT_CELL` , `ENTIRE_TABLE` , `COLUMN_INDEX_VALUE`(value that we are looking for) , `FALSE`(for exact match))

### - **HLOOKUP**
    - Used to look up values in horizontal order
    - It looks up for the first value of selected table and gives the value of selected column index
    
    formula : =HLOOKUP(`INPUT_CELL` , `ENTIRE_TABLE` , `ROW_INDEX_VALUE`(value that we are looking for) , `FALSE`(for exact match))

### - **XLOOKUP**
    - customized search conditions
    - works based on seleted input and ouput arrays condition
    
    formula : = XLOOKUP(`INPUT_CELL` , `SEARCH_ARRAY`(search column) , `RESULT_ARRAY`(output column) )
    
    - Make sure to match input and output column or else add the if not message
    
    formula : = XLOOKUP(`INPUT_CELL` , `SEARCH_ARRAY`(search column) , `RESULT_ARRAY`(output column) , `“any_message”` )
    
    - To lookup with regex values ( unkown )
    
    formula : =XLOOKUP(INPUT_CELL & “*” , `SEARCH_ARRAY`(search column) , `RESULT_ARRAY`(output column) ,  ,  `WILD_CARD_MODE` (look for related values)
    
    - The above formula will show 1st value by deafult

## **Rules for Visualizations**



1. **Define Your Objective**:
    - Clearly define the purpose of your visualization. Understand what insights or messages you want to convey to your audience.
2. **Choose the Right Chart Type**:
    - Select a chart type that best represents your data and effectively communicates your message. Common chart types in Excel include bar charts, line charts, pie charts, scatter plots, and histograms.
3. **Simplify and Declutter**:
    - Keep your visualizations simple and avoid unnecessary clutter. Remove any non-essential elements that do not contribute to the message you want to convey.
4. **Use Appropriate Titles and Labels**:
    - Clearly label your axes, chart titles, and data points. Use descriptive titles and labels that accurately represent the data being visualized.
5. **Provide Context**:
    - Include context or annotations to help your audience understand the significance of the data. This could include trend lines, reference lines, or explanatory text.
6. **Use Color Wisely**:
    - Choose a color scheme that enhances readability and clarity. Use contrasting colors for different data series or categories, and avoid overly bright or distracting colors.
7. **Ensure Data Accuracy**:
    - Double-check your data inputs and calculations to ensure accuracy. Incorrect or misleading data can undermine the credibility of your visualization.
8. **Consider Accessibility**:
    - Make your visualizations accessible to all audiences, including those with color vision deficiencies or visual impairments. Use patterns or textures in addition to color to differentiate data points.
9. **Test and Iterate**:
    - Test your visualization with different audiences or stakeholders to gather feedback. Iterate and refine your visualization based on feedback to improve clarity and effectiveness.
10. **Document Your Process**:
    - Document your data sources, assumptions, and methodology used to create the visualization. This helps maintain transparency and allows others to understand and replicate your analysis.

## Creating Tables

- To create a table - `CREATE TABLE TABLE_NAME ( Column_1 Datatype , Column_2 Datatype …. )`
- For SERIAL Auto-incrementing columns - `CREATE TABLE table_name ( id SERIAL , column1 datatype1 , column2 datatype2 )`
- To set KEY constrains (Must be unique ) - `CREATE TABLE TABLE_NAME ( Column_1 Datatype PRIMARY_KEY , Column_2 Datatype …. )`
- To create a table with UNIQUE values - `CREATE TABLE TABLE_NAME ( Column_1 Datatype , Column_2 Datatype UNIQUE )`
- To create a table with existing table schema -  `CREATE TABLE new_table_name ( LIKE old_table_name INCLUDING ALL )`
- To create table based on SELECT query - `CREATE TABLE new_table_name AS SELECT column1 FROM existing_table_name WHERE condition`
- To create temp table deleted after session ( no view ) -  `CREATE TEMPORARY TABLE TABLE_NAME ( Column_1 Datatype , Column_2 Datatype …. )`

## Insert Table

- You can insert values in two different views
- With schema notation - `INSERT INTO TABLE_NAME ( COLUMN_1 , COLUMN_2 ) VALUES ( VALUE_1 , VALUE_2 )`
- Without schema notation - `INSERT INTO TABLE_NAME VALUES ( VALUE_1 , VALUE_2 )`

## Reading Tables

- Normal read - `SELECT * FROM TABLE_NAME`
- Distinct read - `SELECT DISTINCT COLUMN_NAME FROM TABLE_NAME`
- Conditional read - `SELECT * FROM TABLE_NAME WHERE CONDITION` ( name = ‘some_name’ or age > some_no )
- Limiting read - `SELECT * FROM TABLE_NAME LIMIT SOME_NUMBER` ( +`OFFSET SOME_NUMBER`  → this will neglect some rows and show remaining )
- Ascending or Descending - `SELECT * FROM TABLE_NAME ORDER BY COLUMN_NAME ASC` ( `DESC` )
- For multiple asc or desc arrangements - `SELECT * FROM TABLE_NAME ORDER BY COLUMN_1 ASC , COLUMN_2 DESC`
- Read based on regex ( It works for text/string ) - `SELECT * FROM TABLE_NAME WHERE COLUMN_1 LIKE ‘%a%'` → This will result all the column values containing ‘a’ init
    - **Words ending with 'a' -** `SELECT * FROM table_name WHERE column_1 LIKE '%a'` for multiple condition
    - **Words starting with 'a' -** `SELECT * FROM table_name WHERE column_1 LIKE 'a%'`
    - **Words containing 'a' followed by any character -** `SELECT * FROM table_name WHERE column_1 LIKE '%a_'`
    - **Words containing 'a' preceded by any character -** `SELECT * FROM table_name WHERE column_1 LIKE '_%a'`

## Update Table

- To update a value in a single column based on condition - `UPDATE TABLE_NAME SET COLUMN_NAME = NEW_VALUE WHERE CONDITION` ( for updating all values in column remove where clause )
- To update multiple column values - `UPDATE TABLE_NAME SET COLUMN_NAME = NEW_VALUE , COLUMN_NAME = NEW_VALUE WHERE CONDITION`

- Error on groupby without aggregation on non categorical column : column "emp_data.name" must appear in the GROUP BY clause or be used in an aggregate function

- String aggregation on text - `SELECT designation, STRING_AGG(name, ', ') AS names_concatenated FROM emp_data GROUP BY designation`

## Join Queries

- **Inner Join** - This type of join returns only the rows that have matching values in both tables.
    - Syntax - `SELECT table1.column1,table2.column1 FROM table1 INNER JOIN table2 ON  table1.common_column = table2.common_column;`
    - Syntax - `SELECT t1.column1,t2.column1 FROM table1 t1 INNER JOIN table2 t2 ON  t1.common_column = t2.common_column;`
- **Left Join** (Left Outer Join) - This type of join returns all rows from the left table, and the matched rows from the right table. If no match is found, NULL values are returned for columns from the right table.
    - Syntax - `SELECT * FROM table1 LEFT JOIN table2 ON  table1.common_column = table2.common_column;`
- **Right Join** (Right Outer Join)- This type of join returns all rows from the right table, and the matched rows from the left table. If no match is found, NULL values are returned for columns from the left table.
    - Syntax - `SELECT * FROM table1 RIGHT JOIN table2 ON  table1.common_column = table2.common_column;`
- **Full join** (Full Outer Join)- This type of join returns all rows when there is a match in either the left or right table. If there is no match, the result is NULL on the side that doesn't have a match.
    - Syntax - `SELECT * FROM table1 FULL JOIN table2 ON  table1.common_column = table2.common_column;`
- **Cross Join** - This type of join returns the Cartesian product of the two tables. Each row of the first table is combined with all rows in the second table.
    - Syntax - `SELECT * FROM table1 CROSS JOIN table2 ;`

- The `ROW_NUMBER()` function assigns a unique number to each row according to the ordering specified in the **`OVER`** clause.
    - syntax - `SELECT id, salesperson, amount, sale_date, ROW_NUMBER() OVER (PARTITION BY salesperson ORDER BY sale_date) AS row_num FROM sales;`
- The `RANK()` function assigns a rank to each row within each partition, with the same rank for rows with equal values and gaps in ranks for subsequent rows.
    - syntax - `SELECT id, salesperson, amount, sale_date, RANK() OVER (PARTITION BY salesperson ORDER BY amount DESC) AS rank FROM sales;`
- The `SUM()` function calculates the cumulative sum of the **`amount`** for each row within each partition.
    - syntax - `SELECT id, salesperson, amount, sale_date, SUM(amount) OVER (PARTITION BY salesperson ORDER BY sale_date) AS cumulative_sum FROM sales;`

## **Trigger function**

````python
#update trigger

--Creating a table for employees data
CREATE TABLE employees (
    id SERIAL PRIMARY KEY,
    name VARCHAR(100)
);

select * from employees

--creating table for employees change as logs in employees_log
CREATE TABLE employees_log (
    log_id SERIAL PRIMARY KEY,
    employee_id INT,
    old_name VARCHAR(100),
    new_name VARCHAR(100),
    change_time TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

select * from employees_log


--To make a trigger when if i change the name in employees table it will make a log in employees_log table
CREATE OR REPLACE FUNCTION log_employee_changes()
RETURNS TRIGGER AS $$
BEGIN
    -- Insert a record into employees_log whenever an update occurs
    INSERT INTO employees_log (
        employee_id, old_name, new_name
    )
    VALUES (
        OLD.id, OLD.name, NEW.name
    );
    RETURN NEW;
END;
$$ LANGUAGE plpgsql;

--creating the execution logic how and when it has to be happen
CREATE TRIGGER employee_changes_trigger
AFTER UPDATE ON employees
FOR EACH ROW
EXECUTE FUNCTION log_employee_changes();

-- Insert an initial employee record
INSERT INTO employees (name) VALUES ('John Doe');

select * from employees

-- Update the employee record
UPDATE employees SET name = 'John Smith' WHERE id = 1;

--To check wheather it gets updated or not
select * from employees

-- Check the log table
SELECT * FROM employees_log;
````

````python
#insert trigger

CREATE TABLE employees (
    id SERIAL PRIMARY KEY,
    name VARCHAR(100)
);

--------------------------

CREATE TABLE employees_insert_log (
    log_id SERIAL PRIMARY KEY,
    employee_id INT,
    name VARCHAR(100),
    inserted_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

----------------------

CREATE OR REPLACE FUNCTION log_employee_insertion()
RETURNS TRIGGER AS $$
BEGIN
    -- Insert a record into employees_insert_log whenever an insertion occurs
    INSERT INTO employees_insert_log (
        employee_id, name, inserted_at
    )
    VALUES (
        NEW.id, NEW.name, CURRENT_TIMESTAMP
    );
    RETURN NEW;
END;
$$ LANGUAGE plpgsql;

------------------------


CREATE TRIGGER employee_insertion_trigger
AFTER INSERT ON employees
FOR EACH ROW
EXECUTE FUNCTION log_employee_insertion();


-- Insert an initial employee record
INSERT INTO employees (name) VALUES ('John Doe');


-- Check the insertion log table
SELECT * FROM employees_insert_log;
````

## **Pivot table**

````python
CREATE TABLE Sales (
    Product VARCHAR(50),
    Month VARCHAR(10),
    SalesAmount INT
);

INSERT INTO Sales (Product, Month, SalesAmount) VALUES
('ProductA', 'Jan', 100),
('ProductA', 'Feb', 150),
('ProductB', 'Jan', 200),
('ProductB', 'Feb', 250);
````

````python
#conditional privot using "case"

SELECT
    Product,
    SUM(CASE WHEN Month = 'Jan' THEN SalesAmount ELSE 0 END) AS January,
    SUM(CASE WHEN Month = 'Feb' THEN SalesAmount ELSE 0 END) AS February
FROM
    Sales
GROUP BY
    Product;
````

````python
#group by

SELECT
    Product,
    Month,
    SUM(SalesAmount) AS TotalSales
FROM
    Sales
GROUP BY
    GROUPING SETS ((Product, Month), (Product));
````

````python
#using window function

SELECT
    Product,
    MAX(CASE WHEN Month = 'Jan' THEN SalesAmount END) OVER (PARTITION BY Product) AS January,
    MAX(CASE WHEN Month = 'Feb' THEN SalesAmount END) OVER (PARTITION BY Product) AS February
FROM
    Sales;
````

````python
#create a if else condition for three class classification -> yes or no
# if age is 31-40 | Buys_computer = yes
# if age is <=30 | credit_rating = fair
# if age is >40 | credit_rating = excellent

                                    Age
                        _____________|_____________
                       |             |             |
                      <=30          31-40         >40
                      |              |             |
                    --------         yes          ------------
                   |   STD   |                   |    C_R     |
                  yes        No                 fair        excellent
````

````python
def classify(age, student, credit_rating):
    if age == "<=30":
        if student == "yes":
            return "yes"
        else:
            return "no"
    elif age == "31-40":
        return "yes"
    elif age == ">40":
        if credit_rating == "fair":
            return "yes"
        elif credit_rating == "excellent":
            return "no"

# Allow user input for classification
user_age = input("Enter age (<=30, 31-40, >40): ")
user_student = input("Are you a student? (yes, no): ")
user_credit_rating = input("Enter credit rating (fair, excellent): ")

result = classify(user_age, user_student, user_credit_rating)
print(f"Buys_computer: {result}")
````

````python
#features -> independent column
#labels -> dependent column

x = data.drop(['Buys_computer','Unnamed: 0'],axis=1)
y = data.Buys_computer
````

````python
x
````

````python
y
````

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

x.age = le.fit_transform(x.age)
x.income = le.fit_transform(x.income)
x.student = le.fit_transform(x.student)
x.Credit_rating = le.fit_transform(x.Credit_rating)

y = le.fit_transform(y)
````

````python
x.values
````

````python
y
````

````python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier() #algo load

model.fit(x.values,y) #algo data pass (training the ml model)
````

````python
#userinput for classification
user_age = 2 #'>40'
user_income = 2 #'medium'
user_student =1 # 'yes'
user_credit_rating =1 #'fair'

result = model.predict([[user_age,user_income,user_student,user_credit_rating]])

if result == 1:
  print("they will buy computer")
else:
  print("they will not buy computer")
````

````python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,8))

plot_tree(model,feature_names=x.columns,filled=True)

plt.show()
````
